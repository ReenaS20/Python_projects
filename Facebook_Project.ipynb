```python
fb_age_friend=fb_raw_data[["age","friend_count"]]
fb_age_friend.head
print("Sliced data by coloumns")
fb_age_friend.shape

fb_age_friend_g=fb_age_friend.groupby(['age'])
fb_age_friend_g

fb_age_friend_g_sum= fb_age_friend_g.sum()
fb_age_friend_g_sum=fb_age_friend_g_sum.reset_index()

fb_age_friend_g_sum.shape
fb_age_friend_g_sum.head

fb_age_friend_g_sum.set_index('age')

fb_age_friend_g_sum.plot.bar(x='age',y='friend_count',rot=0,figsize=(35,12))
plt.xlabel('Age')
plt.ylabel('Friends count age wise in millions')
plt.title('Bar chart sum of friends  age wise')
```
